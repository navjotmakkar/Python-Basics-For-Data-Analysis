```python
import seaborn as sns
```

```python
df=sns.load_dataset('tips')
df.head()
```

## a.Continuous Plot

### 1. Correlation with Heatmap

```python
df.corr()
```

```python
sns.heatmap(df.corr())
```

### 2. Join Plot

```python
sns.jointplot(x='tip', y='total_bill', data=df, kind='hex')
```

```python
sns.jointplot(x='tip', y='total_bill', data=df, kind='reg') #shows pdf and regression line
```

### 3. Pair Plot

```python
sns.pairplot(df) #scatterplot
```

```python
sns.pairplot(df, hue='sex')
```

### 4. Dist Plot

```python
sns.distplot(df['tip'])
```

```python
sns.distplot(df['tip'], kde=False, bins=10)
```

## b. Categorical Plot

### 1. Count Plot

```python
sns.countplot('smoker', data=df)
```

```python
sns.countplot(y='day', data=df)
```

### 2. Bar Plot

```python
sns.barplot(x='smoker', y='total_bill', data=df)
```

### 3. Box Plot

```python
sns.boxplot(x='day', y='total_bill', data=df, palette='rainbow')
```

```python
sns.boxplot(data=df, orient='y')
```

```python
sns.boxplot(x='day', y='total_bill', data=df, hue='smoker')
```

### 4.Violin Plot

```python
sns.violinplot(x='day', y='total_bill', data=df, palette='rainbow')
```
