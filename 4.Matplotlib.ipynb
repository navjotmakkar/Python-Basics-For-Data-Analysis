## Matplotlib

```python
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
import numpy as np
```

```python
x=np.arange(0,10)
y=np.arange(11,21)
```

```python
plt.scatter(x,y,c='g')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Graph')
plt.savefig('Test.png')
```

```python
y=x*x
```

```python
plt.plot(x,y,'r*-')
```

```python
plt.plot(x,y,'r*',linestyle='dashed',linewidth=2, markersize=12)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('graph')
```

```python
plt.subplot(2,2,1)
plt.plot(x,y,'r')
plt.subplot(2,2,2)
plt.plot(x,y,'go-')
plt.subplot(2,2,3)
plt.plot(x,y,'bo')
plt.subplot(2,2,4)
plt.plot(x,y,'y*-')
```

```python
x=np.arange(0,4*np.pi,0.1)
y_sin=np.sin(x)
plt.plot(x,y)
plt.title('sine wave form')
```

```python
y_cos=np.cos(x)
plt.subplot(2,1,1)
plt.title('sine')
plt.plot(x,y_sin)
plt.subplot(2,1,2)
plt.title('cosine')
plt.plot(x,y_cos)
```

### Bar Graph

```python
x1=[2,8,10]
y1=[13,16,9]

x2=[2,9,11]
y2=[6,15,7]

plt.bar(x1,y1,align='center')
plt.bar(x2,y2,color='g')
plt.title('Bar graph')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
```

### Histogram

```python
a=np.random.randint(2,60,15)
plt.hist(a, bins=15)
plt.title('histogram')
```

### Box Plot

```python
data=[np.random.normal(2,std,100) for std in range(1,4)]
plt.boxplot(data, vert=True, patch_artist=True)
```

### Pie Chart

```python
labels = 'Python', 'C++', 'Ruby', 'Java'
sizes = [215, 130, 245, 210]
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
explode = (0.4, 0, 0, 0)  # exploding 0nly 1st slice

plt.pie(sizes, explode=explode, labels=labels, colors=colors,
autopct='%1.1f%%', shadow=False)

plt.axis('equal')
plt.show()
```
